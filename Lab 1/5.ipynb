```python
import torch
import torch.nn as nn
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("house_price_full.csv")
```

```python
X = df[['bedrooms', 'sqft_living']].values
y = df[['price']].values
```

```python
scaler_X = StandardScaler()
scaler_y = StandardScaler()

X = scaler_X.fit_transform(X)
y = scaler_y.fit_transform(y)
```

```python
X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.float32)
```

```python
model = nn.Sequential(
    nn.Linear(2, 8),
    nn.ReLU(),
    nn.Linear(8, 1)
)
```

```python
loss_fn = nn.MSELoss()
```

```python
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
for epoch in range(1000):
    y_pred = model(X)
    loss = loss_fn(y_pred, y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
test_house = torch.tensor([[3, 2000]], dtype=torch.float32)
test_house = scaler_X.transform(test_house)
test_house = torch.tensor(test_house, dtype=torch.float32)

predicted_price = model(test_house)
predicted_price = scaler_y.inverse_transform(predicted_price.detach().numpy())

print(predicted_price)
```
