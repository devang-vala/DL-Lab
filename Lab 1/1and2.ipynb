```python
import torch
```

### 1 Question

```python
# From Python list
a = torch.tensor([1, 2, 3])
print("a:", a)

# Zeros tensor
b = torch.zeros(3)
print("b:", b)

# Ones tensor
c = torch.ones(2, 2)
print("c:\n", c)

# Random tensor
d = torch.rand(2, 3)
print("d:\n", d)

# Range tensor
e = torch.arange(0, 10, 2)
print("e:", e)
```

```python
x = torch.tensor([10, 20, 30])
y = torch.tensor([1, 2, 3])

print("x + y =", x + y)
print("x - y =", x - y)
print("x * y =", x * y)
print("x / y =", x / y)
```

```python
A = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])

B = torch.tensor([10, 20, 30])

print("A:\n", A)
print("B:", B)

print("A + B (Broadcasting):\n", A + B)
```

```python
T = torch.tensor([[10, 20, 30],
                  [40, 50, 60]])

print("T:\n", T)

print("First row:", T[0])
print("First column:", T[:, 0])
print("Element at (1,2):", T[1, 2])
```

```python
t = torch.arange(1, 13)
print("Original:", t)

t2 = t.view(3, 4)
print("Reshaped (3x4):\n", t2)

t3 = t2.reshape(2, 6)
print("Reshaped (2x6):\n", t3)
```

```python
x = torch.tensor(2.0, requires_grad=True)

y = x**2 + 3*x + 1
print("y =", y)

y.backward()  # computes dy/dx

print("Gradient (dy/dx) =", x.grad)
```

```python
x = torch.tensor([1.0, 2.0, 3.0], requires_grad=True)

y = (x * x).sum()  # sum(x^2)
y.backward()

print("x:", x)
print("Gradient:", x.grad)
```

### 2 Question - Linear Algebra Operations

```python
# Create Matrices (Tensors)
A = torch.tensor([[1., 2.],
                  [3., 4.]])

B = torch.tensor([[5., 6.],
                  [7., 8.]])

print("Matrix A:\n", A)
print("Matrix B:\n", B)
```

```python
# Matrix Addition & Subtraction
print("A + B:\n", A + B)
print("A - B:\n", A - B)
```

```python
# Scalar Multiplication
k = 2
print("k * A:\n", k * A)
```

```python
# Element-wise Multiplication
print("A * B (element-wise):\n", A * B)
```

```python
# Matrix Multiplication
print("A @ B (matrix multiplication):\n", A @ B)
```

```python
# Transpose
print("Transpose of A:\n", A.T)
```

```python
# Determinant
det_A = torch.det(A)
print("Determinant of A:", det_A.item())
```

```python
# Inverse of Matrix
inv_A = torch.inverse(A)
print("Inverse of A:\n", inv_A)
```

```python
# Identity Matrix
I = torch.eye(2)
print("Identity Matrix:\n", I)
```

```python
# Verify A × A⁻¹ = I
print("A @ inv(A):\n", A @ inv_A)
```

```python
# Dot Product (Vectors)
v1 = torch.tensor([1., 2., 3.])
v2 = torch.tensor([4., 5., 6.])

dot = torch.dot(v1, v2)
print("Dot product:", dot.item())
```

```python
# Norm (Magnitude of Vector)
norm_v1 = torch.norm(v1)
print("Norm of v1:", norm_v1.item())
```

```python
# Trace of Matrix
trace_A = torch.trace(A)
print("Trace of A:", trace_A.item())
```

```python
# Rank of Matrix
rank_A = torch.linalg.matrix_rank(A)
print("Rank of A:", rank_A.item())
```

```python
# Eigenvalues & Eigenvectors
eig_vals, eig_vecs = torch.linalg.eig(A)

print("Eigenvalues:\n", eig_vals)
print("Eigenvectors:\n", eig_vecs)
```

```python
# Solve Linear Equations (Ax = b)
A = torch.tensor([[3., 1.],
                  [1., 2.]])

b = torch.tensor([[9.],
                  [8.]])

x = torch.linalg.solve(A, b)

print("Solution x:\n", x)
```
